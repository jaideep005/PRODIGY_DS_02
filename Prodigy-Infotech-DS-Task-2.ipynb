### CHECKING FOR DUPLICATE,MISSING AND CLEANING OF DATA

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop('Cabin', axis=1, inplace=True)
```

```python
data.head()
```

```python
data['Age'] = data.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
data.head(10)
```

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.head(10)
```

### EXPLOARING DATA ANALYSIS (EDA)

```python
plt.figure(figsize=(6,3))
sns.histplot(data["Age"],kde=True)
plt.title("Age Distributed")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=data,x="Sex",hue="Sex")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python

plt.figure(figsize=(6,3))
sns.scatterplot(data=data,x="Age",y="Fare", hue="Age")
plt.title("Scatter plot of Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
plt.hist(data['Ticket'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Ticket')
plt.xlabel('Tickets')
plt.ylabel('Count')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='Fare', hue='Sex', multiple="stack", palette="pastel", kde=False)
plt.title('Distribution of Ticket Fares by Sex')
plt.ylabel('Count')
plt.xlabel('Fare')
plt.show()
```
